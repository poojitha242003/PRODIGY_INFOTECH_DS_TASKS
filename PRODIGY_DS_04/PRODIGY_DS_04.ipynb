```python
import pandas as pd
import matplotlib.pyplot as plt
from textblob import TextBlob
```

```python
df = pd.read_csv("sentiment_data.csv")
df.head()
```

```python
df.info()
```

```python
def get_sentiment(text):
    analysis = TextBlob(str(text))
    if analysis.sentiment.polarity > 0:
        return "Positive"
    elif analysis.sentiment.polarity < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
df.columns
```

```python
df = pd.read_csv("sentiment_data.csv", header=None)
df.head()
```

```python
df.columns = ["ID", "Topic", "Label", "Text"]
df.head()
```

```python
df["Predicted_Sentiment"] = df["Text"].apply(get_sentiment)
df.head()
```

```python
df["Predicted_Sentiment"].value_counts()
```

```python
df["Predicted_Sentiment"].value_counts().plot(kind="bar")
plt.title("Sentiment Distribution on Social Media Data")
plt.xlabel("Sentiment")
plt.ylabel("Number of Posts")
plt.show()
```
