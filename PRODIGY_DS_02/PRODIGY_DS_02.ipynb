```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
# Load CSV (since it's in the same folder as notebook)
df = pd.read_csv("StudentsPerformance.csv")

# Check first few rows
df.head()

```

```python
# Check the shape of the dataset (rows, columns)
df.shape
# Check column names
df.columns
# Get basic info (data types, null values)
df.info()
# Quick statistics for numeric columns
df.describe()
```

```python
# Histogram for math scores
plt.figure(figsize=(8,5))
sns.histplot(df['math score'], bins=10, kde=True, color='skyblue')
plt.title("Distribution of Math Scores")
plt.xlabel("Math Score")
plt.ylabel("Count")
plt.show()
```

```python
# Count of students by gender
plt.figure(figsize=(6,4))
sns.countplot(x='gender', data=df, palette='pastel')
plt.title("Number of Students by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Histogram for reading scores
plt.figure(figsize=(8,5))
sns.histplot(df['reading score'], bins=10, kde=True, color='lightgreen')
plt.title("Distribution of Reading Scores")
plt.xlabel("Reading Score")
plt.ylabel("Count")
plt.show()
```

```python
# Bar chart for parental level of education
plt.figure(figsize=(10,5))
sns.countplot(x='parental level of education', data=df, palette='viridis', order=df['parental level of education'].value_counts().index)
plt.title("Number of Students by Parental Level of Education")
plt.xlabel("Parental Level of Education")
plt.ylabel("Count")
plt.xticks(rotation=45)  # Rotate x labels for better visibility
plt.show()
```
